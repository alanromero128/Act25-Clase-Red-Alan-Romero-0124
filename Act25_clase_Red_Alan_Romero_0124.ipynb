# Clase Red
- Atributos:  (nombre, tipo, velocidad)
- Funciones: leerdatos(), mostrardatos()
- Objetos : (red1, red2)

![link text](https://raw.githubusercontent.com/alanromero128/Act25-Clase-Red-Alan-Romero-0124/refs/heads/main/infoclaseeRed.png.png)

```python
class Red:
    # Constructor con los 3 atributos solicitados
    def __init__(self, nombre, tipo, velocidad):
        self.nombre = nombre
        self.tipo = tipo
        self.velocidad = velocidad

    # Función 1: Leer datos
    def leer_datos(self):
        print(f"\n--- Ingresando datos para la red ---")
        self.nombre = input("Nombre de la red: ")
        self.tipo = input("Tipo de red: ")
        self.velocidad = input("Velocidad de la red: ")

    # Función 2: Mostrar datos
    def mostrar_datos(self):
        print(f"\n[DATOS DE LA RED]")
        print(f"Nombre: {self.nombre} | Tipo: {self.tipo} | Velocidad: {self.velocidad}")

# --- Creación de los 2 Objetos ---
# Crear red1 (vacía o con datos iniciales temporales, se pedirá input)
red1 = Red("", "", "") # Inicializar con valores vacíos para que leer_datos los sobrescriba
red2 = Red("", "", "") # Inicializar con valores vacíos
# Pedir y mostrar datos para Red 1
print("\n--- Configurando Red 1 ---")
red1.leer_datos()

# Pedir y mostrar datos para Red 2
print("\n--- Configurando Red 2 ---")
red2.leer_datos()

red1.mostrar_datos()
red2.mostrar_datos()

print("Programa terminado")
print("Programa Realizado por Alan Romero NL = 47 ")
```
